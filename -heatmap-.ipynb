```python
import streamlit as st
import pandas as pd
import folium
from folium import plugins
from folium.plugins import HeatMap

# Load the CSV file
data = pd.read_csv('berlin_data.csv')

# Create a folium map centered on the mean coordinates
first_map = folium.Map(location=[data['lat'].mean(), data['lon'].mean()], zoom_start=12)

first_map
```

```python
heat_data = [[row['lat'],row['lon']] for index, row in data.iterrows()]

heat_data
```

```python
heat_map = HeatMap(heat_data)
map_2 = heat_map.add_to(first_map)
map_2
```

```python
map_2.save('heatmap.html')
```
