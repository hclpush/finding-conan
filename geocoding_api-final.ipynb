```python
# define a `coordinates` function:
# - taking an address as parameter
# - returning a latitude and a longitude

def coordinates(location):
    url = "https://nominatim.openstreetmap.org/search"
    params = {
        'q': location,
        'format': 'json'
    }
    response = requests.get(url, params=params).json()
    return response
```

```python
selected_time = 'Early Morning (04-07)'

def compute_radius(selected_time):
    # Define the mapping of time ranges to radius values
    radius_mapping = {
        'Early Morning (04-07)': 10,
        'Morning (07-11)': 15,
        'Noon(11-14)': 20,
        'Afternoon (14-17)': 25,
        'Evening (17-21)': 30,
        'Night (21-23)': 25,
        'Midnight (23-04)': 15,
        
    }



    # Get the radius for the selected time range
    radius = radius_mapping[selected_time]
    return radius

radius = compute_radius(selected_time)
radius
```
